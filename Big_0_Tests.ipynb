```python
## o(n) test - essentially worst case scenario for the run.
def print_test(n):
    # start of function
    for i in range(n):
        # iteration triggers for the n items
        print(i)
        # print each iteration item
```

```python
print_test(10)
```

```python
## o(n^2) test - nested loops.
def print_test_2(n):
    # start of the function
    for i in range(n):
        # outer loop for n items
        for j in range(n):
            # inner loop for the n items
            print(i, j)
            # print both outer loop and inner loop items causing n^2 complexity
```

```python
print_test_2(5)
```

```python
# Big O :o(log n) test - basically the number of times we have to divide or run to get to the number we want example lon 2^8=3 or 2^3 is 8
def print_test_3(n):
    # start of function
    i = 1
    while i < n:
        # loop that runs while i is less than n
        print(i)
        # print the current value of i
        i *= 2
        # double the value of i each iteration
print_test_3(16)


```

Notes : interesting concept is taking a list removing and adding to the end will always be 0(1) as only oeration performed. while adding removing anywhere else is 0(n) as other items will need to be reordered for the operation touching all items in the list.
